```python
import numpy as np
import matplotlib.pyplot as plt
# Number of sample points
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(-2.0, 4.0, N)
y = (2*x + 0j)**(2/3.0) + 0j
yf = np.fft.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(x, y)

plt.subplot(212)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```
